```python
df.dropna(subset = ["Club", "Position", "Joined", "Contract Valid Until"], inplace = True) #drop all the rows where "Club", "Position", "Joined", "Contract Valid Until" is null directly from data
df.shape
```

```python
df["Loaned From"].fillna("None", inplace = True) #Replace Null values in "Loaned From" column with "None"
df["Release Clause"].fillna("€0", inplace = True) #Replace Null values in "Release Claus" column with "€0"
missing_values_count = df.isnull().sum()
missing_values_count[["Loaned From", "Release Clause"]]
```

```python
GKdf = df[df["Position"] == "GK"] #Create a new dataset where "Position" = "GK"
GKdf.shape
```

```python
GKdf.head()
```

```python
GKdf.drop(["LS","ST","RS"], axis = 1) #Show the dataset without "LS", "ST", "RS" column
```

```python
del GKdf["LS"] #Remove "LS" column from dataset
GKdf.shape
```

```python
Gkdf = GKdf.drop(['ST', 'RS', 'LW', 'LF', 'CF', 'RF', 'RW',
       'LAM', 'CAM', 'RAM', 'LM', 'LCM', 'CM', 'RCM', 'RM', 'LWB', 'LDM',
       'CDM', 'RDM', 'RWB', 'LB', 'LCB', 'CB', 'RCB', 'RB'], axis = 1, inplace = True) #Remove the list of columns from dataset
GKdf.shape
```

```python
GK_missing_values_count = GKdf.isnull().sum()
GK_missing_values_count.sum()
```
